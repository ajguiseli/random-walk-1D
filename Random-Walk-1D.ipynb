```python
import matplotlib.pyplot as plt
import numpy as np

N = 1000
x = np.zeros(N)
x[0] = 0

for i in range(N-1):
    x[i+1] = x[i] + np.random.randn()

t = np.arange(N)

plt.figure()
plt.plot(t, x, marker='o')
plt.xlabel("Passo (i)")
plt.ylabel("Posição x[i]")
plt.title("Random Walk 1D – visualização passo a passo")
plt.show()
```
